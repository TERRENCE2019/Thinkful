```
#My data is called Jobs so I named "Jobs" my dataframe (change the name of yours relative to data)
Jobs=pd.read_csv("https://raw.githubusercontent.com/TERRENCE2019/Thinkful/master/nyc-jobs.csv")
```

```

#Rename Dataframe to not overwrite original
#Execute Code
#Pick two columns
Jobs2 = Jobs.plot.line(x='# Of Positions', y='Salary Range To')

plt.title('Line Plot')
plt.ylabel('Salary Range To')
plt.xlabel('# of positions')

plt.show()
```
